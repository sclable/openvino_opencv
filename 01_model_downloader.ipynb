### Model Downloader
Download trained models from OpenVINO model zoo repository. 
It consists of two types of model repos: 
* Intel trained models which are in already IR formats [1] and 
* Public models which are in their native framework's formats [2] (.pb, .pth, .caffe etc) 

[1] https://docs.openvinotoolkit.org/latest/omz_models_intel_index.html 
 
[2] https://docs.openvinotoolkit.org/latest/omz_models_public_index.html

```python
import os
```

```python
# Parameters required for model downloading
MODEL_NAME = 'face-detection-retail-0004' # 'squeezenet1.0' #"face-detection-retail-0004" 
OUT_DIR = os.path.join(os.getcwd(), "models")
PRECISION = 'FP16' # other options are: FP32, FP16 or INT8
```

```python
# Create a directory to save models 
if not os.path.exists(OUT_DIR):
    os.mkdir(OUT_DIR)
```

```python
# Run the model_downloader script
!/opt/intel/openvino/deployment_tools/tools/model_downloader/downloader.py \
--name {MODEL_NAME} \
--precisions {PRECISION} \
--output_dir {OUT_DIR}
```
